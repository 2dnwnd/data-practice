## COVID-19 백신 종류별 접종 인구

```python
day = [1,2,3] # 1일,2일,3일
az = [2,4,8] # 아스트라제네카
pfizer = [5,1,3]
moderna = [1,2,5] 

plt.plot(day, az)
plt.plot(day, pfizer)
plt.plot(day, moderna)
```

```python
plt.plot(day, az, label='az')
plt.plot(day, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(day, moderna, label='moderna', marker='s', ls='-.')

plt.legend()
```

```python
plt.plot(day, az, label='az')
plt.plot(day, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(day, moderna, label='moderna', marker='s', ls='-.')

plt.legend(ncol=2)
```
