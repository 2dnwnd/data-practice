```python
import matplotlib.pyplot as plt

years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.plot(years,gdp,color='green',marker='o',linestyle='solid')

plt.title('GDP per capita')

plt.ylabel('dollars')
plt.savefig('gdp_per_capita.png', dpi = 600)
plt.show()
```

```python
x = [x for x in range(-20,20)]
y1 = [2*t for t in x]
y2 = [t**2+5 for t in x]
y3 = [-t**2-5 for t in x]
#ㅂ빨간색점선, 녹색 실선과 세모기호 , 파란색별표와 점선
plt.plot(x,y1,'r--',x,y2,'g^-',x,y3,'b*:')
plt.axis([-30,30,-30,30])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [x for x in range(20)]
y = [x**2 for x in range(20)]
z = [x**3 for x in range(20)]
r = [2**x for x in range(20)]
print(r)
plt.plot(x,x,label='linear')
plt.plot(x,y,label='quadratic')
plt.plot(x,z,label='qubic')
plt.plot(x,r,label='power')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('My Plot')
plt.legend()
plt.axis([0,2,0,8])
plt.show()
```

```python
import math

x = []
y1 =[]
y2 = []

for angle in range(360):
    x.append(angle)
    y1.append(math.sin(math.radians(angle)))
    y2.append(math.cos(math.radians(angle)))
    
plt.plot(x,y1,x,y2,'r-')
plt.title('SINE WAVE')
plt.show()
```

```python
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.bar(range(len(years)),gdp)
plt.title('GDP per capita')
plt.ylabel('dollars')

plt.xticks(range(len(years)),years)
plt.show()
```

```python
years = [1965,1975,1985,1995,2005,2015]
ko = [130,650,2450,11600,17790,27250]
jp = [890,5120,11500,42130,40560,38780]
ch = [100,200,290,540,1760,7940]

#x_range = range(len(years)) 
#range함수는 실수값을 더할 수 없다
x_range = np.arange(len(years))
plt.bar(x_range+0.0, ko, width=0.25, label='korea')
plt.bar(x_range+0.3, jp, width=0.25, label='japan')
plt.bar(x_range+0.6, ch, width=0.25, label='china')
plt.xticks(range(len(years)),years)
plt.title('GDP per capita')
plt.ylabel('dollars')
plt.legend()
plt.show()
```
