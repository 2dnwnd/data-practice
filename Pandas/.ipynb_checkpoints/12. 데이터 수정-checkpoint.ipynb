# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고'})
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['학교'] = df['학교'] + '등학교'
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail'
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass'
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합 컬럼 삭제
```

```python
df.drop(columns=['국어','수학'])
```

## Row 삭제

```python
df.drop(index='4번')
```

```python
filt = df['수학'] < 80 # 수학점수 80점 미만학생 필터링
df[filt]
```
